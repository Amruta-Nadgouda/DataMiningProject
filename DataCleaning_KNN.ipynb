```python
totalDF = totalDF.fillna(0)
totalDF
```

```python
def probViolation(row):
    probVal = 0
    vehCol = 0
    strName = 0
    locTime = 0
    
    if (row['Vehicle Color'] == "WHITE" or row['Vehicle Color'] =="GY" or row['Vehicle Color'] =="WH" or row['Vehicle Color'] =="BK" or row['Vehicle Color'] =="BLACK" or row['Vehicle Color'] =="GREY"):
        vehCol = 1
    
    if (row['Street Name'] == "Broadway" or row['Street Name'] =="3rd Ave" or row['Street Name'] =="5th Ave"):
        strName = 1
    
    if (row['Violation Time'] == "0836A" or row['Violation Time'] =="1136A" or row['Violation Time'] == "1140A" or row['Violation Time'] =="0936A"):
        locTime = 1
        
    probVal = (vehCol + strName + locTime)/3
    
    if probVal > 0.3:
        probVal = 1
    else:
        probVal = -1
    
    return probVal
```

```python
totalDF['y'] = totalDF.apply(probViolation, axis=1)
```

```python
totalDF['y'].value_counts()
```

```python
y = totalDF['y']
y
```

```python
subDF = pd.DataFrame({'Vehicle Color':totalDF['Vehicle Color'],'Street Name':totalDF['Street Name'],'Violation Time':totalDF['Violation Time']})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
subDF.head()
```

```python
le.fit(subDF['Vehicle Color'].astype(str))
subDF['Vehicle Color']= le.transform(subDF['Vehicle Color'].astype(str))
subDF
#subDF['Vehicle Color'] = le.transform(subDF['Vehicle Color']                                    
```

```python
le.fit(subDF['Street Name'].astype(str))
subDF['Street Name']= le.transform(subDF['Street Name'].astype(str))
subDF
```

```python
le.fit(subDF['Violation Time'].astype(str))
subDF['Violation Time']= le.transform(subDF['Violation Time'].astype(str))
subDF
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve,auc
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(subDF, 
                                                    y,
                                                    test_size=0.20,
                                                    random_state=42)
```

```python
classifier = KNeighborsClassifier()
```

```python
classifier.fit(X_train,y_train)
predTest = classifier.predict(X_test)
```

```python
print('Accuracy: {:.2%}\n'.format(metrics.accuracy_score(y_test, predTest)))
print('F1 Score: {:.2%}\n'.format(metrics.f1_score(y_test, predTest, average='macro')))
print('Precision: {:.2%}\n'.format(metrics.precision_score(y_test, predTest, average='macro')))
print('Recall: {:.2%}\n'.format(metrics.recall_score(y_test, predTest, average='macro')))
```
