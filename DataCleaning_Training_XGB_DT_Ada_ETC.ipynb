```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import zipfile

def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
```

```python
with zipfile.ZipFile("nyc-parking-tickets.zip","r") as zip_ref:
     zip_ref.extractall("parkingtickets")
```

```python
dataf2014 = pd.read_csv('parkingtickets/Parking_Violations_Issued_-_Fiscal_Year_2014__August_2013___June_2014_.csv')
dataf2014.head()
```

```python
dataf2015 = pd.read_csv('parkingtickets/Parking_Violations_Issued_-_Fiscal_Year_2015.csv')
dataf2015.head()
```

```python
dataf2016 = pd.read_csv('parkingtickets/Parking_Violations_Issued_-_Fiscal_Year_2016.csv')
dataf2016.head()
```

```python
dataf2017 = pd.read_csv('parkingtickets/Parking_Violations_Issued_-_Fiscal_Year_2017.csv')
dataf2017.head()
```

```python
dataf2014DF = pd.DataFrame(dataf2014)
dataf2014DF.head()
```

```python
dataf2015DF = pd.DataFrame(dataf2015)
dataf2015DF.head()
```

```python
dataf2016DF = pd.DataFrame(dataf2016)
dataf2016DF.head()
```

```python
dataf2017DF = pd.DataFrame(dataf2017)
dataf2017DF.head()
```

```python
totalDF = dataf2014DF
totalDF.shape
```

```python
# concatenating the dataframes dataf2014DF and dataf2015DF
totalDF = pd.concat([totalDF,dataf2015DF], ignore_index=True)
totalDF.shape
```

```python
# concatenating the dataframes dataf2014DF and dataf2015DF and dataf2016DF
totalDF = pd.concat([totalDF,dataf2016DF], ignore_index=True)
totalDF.shape
```

```python
# we are not concatenating dataf2017DF as it has only 43 features and does not satisfy the fetaure requirements
# totalDF = pd.concat([totalDF,dataf2017DF], ignore_index=True, axis = 0)
# totalDF.shape
```

```python
totalDF.head()
```

```python
# filling the unidentified values with zero to make the data usable and also to maintain consistency across the dataframe.
totalDF = totalDF.fillna(0)
totalDF
```

```python
# below function checks for the features like vehicle color, street name, violation time and 
# calculates the probability of a person obtaining the ticket and hence makes predictions on the same.
def probViolation(row):
    probVal = 0
    vehCol = 0
    strName = 0
    locTime = 0
    
    if (row['Vehicle Color'] == "WHITE" or row['Vehicle Color'] =="GY" or row['Vehicle Color'] =="WH" or row['Vehicle Color'] =="BK" or row['Vehicle Color'] =="BLACK" or row['Vehicle Color'] =="GREY"):
        vehCol = 1
    
    if (row['Street Name'] == "Broadway" or row['Street Name'] =="3rd Ave" or row['Street Name'] =="5th Ave"):
        strName = 1
    
    if (row['Violation Time'] == "0836A" or row['Violation Time'] =="1136A" or row['Violation Time'] == "1140A" or row['Violation Time'] =="0936A"):
        locTime = 1
        
    probVal = (vehCol + strName + locTime)/3
    
    if probVal > 0.3:
        probVal = 1
    else:
        probVal = -1
    
    return probVal
```

```python
# applying the above function to that complete dataframe
totalDF['y'] = totalDF.apply(probViolation, axis=1)
```

```python
totalDF['y'].value_counts()
```

```python
y = totalDF['y']
y
```

```python
# Adding only important features into a dataframe
subDF = pd.DataFrame({'Vehicle Color':totalDF['Vehicle Color'], 'Street Name':totalDF['Street Name'], 'Violation Time':totalDF['Violation Time']})
subDF
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, ExtraTreesClassifier
```

```python
le = LabelEncoder()
le.fit(subDF['Vehicle Color'].astype(str))
subDF['Vehicle Color'] = le.transform(subDF['Vehicle Color'].astype(str))
```

```python
le.fit(subDF['Street Name'].astype(str))
subDF['Street Name'] = le.transform(subDF['Street Name'].astype(str))
```

```python
le.fit(subDF['Violation Time'].astype(str))
subDF['Violation Time'] = le.transform(subDF['Violation Time'].astype(str))
```

```python
subDF
```

```python
# Split data into train-test
X_train, X_test, y_train, y_test = train_test_split(subDF, 
                                                    y,
                                                    test_size=0.20,
                                                    random_state=42)
```

## Support Vector Classifier

```python
classifier =  SVC(C=5, gamma = 0.5)
```

```python
classifier.fit(X_train, y_train)
predTest = classifier.predict(X_test)
```

```python
print('Accuracy: {:.2%}\n'.format(metrics.accuracy_score(y_test, predTest)))
print('F1 Score: {:.2%}\n'.format(metrics.f1_score(y_test, predTest, average='macro')))
print('Precision: {:.2%}\n'.format(metrics.precision_score(y_test, predTest, average='macro')))
print('Recall: {:.2%}\n'.format(metrics.recall_score(y_test, predTest, average='macro')))
```

SVC took long to converge may be because of the large dataset. Therefore we will not be using SVC as one of our classifiers.

## XGBoost Classifier

```python
print('Accuracy: {:.2%}\n'.format(metrics.accuracy_score(y_test, predTest_xgb)))
print('F1 Score: {:.2%}\n'.format(metrics.f1_score(y_test, predTest_xgb, average='macro')))
print('Precision: {:.2%}\n'.format(metrics.precision_score(y_test, predTest_xgb, average='macro')))
print('Recall: {:.2%}\n'.format(metrics.recall_score(y_test, predTest_xgb, average='macro')))
```
